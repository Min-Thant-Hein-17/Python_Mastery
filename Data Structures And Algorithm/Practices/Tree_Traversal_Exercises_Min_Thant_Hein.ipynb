```python
# Task: Build a binary family tree and implement preorder, inorder, postorder traversals

class FamilyMember:
    def __init__(self, name, age, title):
        self.name = name
        self.age = age
        self.title = title
        self.left = None
        self.right = None

# Example: Create root member
root = FamilyMember("Arthur", 60, "King")

# TODO: Add children and grandchildren nodes

root.left = FamilyMember("Lancelot", 35, "Knight")
root.right = FamilyMember("Gunivere", 32, "Queen")
root.left.left = FamilyMember("Galahad", 14, "Squire")
root.right.right = FamilyMember("Elaine", 18, "Princess")

# TODO: Implement tree traversals
def preorder(node):
    if node:
        print(f"{node.title} {node.name}")
        preorder(node.left)
        preorder(node.right)

def inorder(node):
    if node:
      inorder(node.left)
      print(f"{node.title} {node.name}")
      inorder(node.right)

def postorder(node):
    if node:
      postorder(node.left)
      postorder(node.right)
      print(f"{node.title} {node.name}")

# Call the traversals
print("Preorder Traversal:")
preorder(root)
print("\nInorder Traversal:")
inorder(root)
print("\nPostorder Traversal:")
postorder(root)

```

When searching with find_animal_path(entrance, "Penguin Zone", []), the output should be:

['Main Gate', 'Bird Sanctuary', 'Penguin Zone']



```python
# Task: Represent zoo as a tree and search for animal locations using DFS

class ZooZone:
    def __init__(self, name):
        self.name = name
        self.children = []

# Build tree
entrance = ZooZone("Main Gate")
# TODO: Add zones like Reptile House, Bird Sanctuary, etc.   #Establishing Relationships between Zoo Zones
reptile_house = ZooZone("Reptile House")
bird_sanctuary = ZooZone("Bird Sanctuary")
snake_pit = ZooZone("Snake Pit")
penguin_zone = ZooZone("Penguin Zone")
lion_zone = ZooZone("Lion Zone")

#Building the Zoo Tree Structure for indicating the pathways

entrance.children.append(reptile_house)
entrance.children.append(bird_sanctuary)
reptile_house.children.append(snake_pit)
bird_sanctuary.children.append(penguin_zone)

# TODO: DFS to find all paths to animals (leaf zones)
def list_paths(node, path):         # Implementing recursive traversal here
    path.append (node.name)
    if not node.children:             #Checking whether current node has any children or not
        print(" -> ".join(path))      #If there is no child, the condition is true, printing the entire path that has been constructed
    else:
        for child in node.children:
            list_paths(child, path[:])  #Creating a copy of the current path list # Crucial step for recursion


# TODO: Find a specific animal's path
def find_animal_path(node, target, path): # Returning the path we want to find to the target zone
    path.append (node.name)
    if node.name == target:  #Checking the Target
        return path
    for child in node.children:  #Iterating Through Children Nodes
        result = find_animal_path(child, target, path[:])  #Recursive Search
        if result:                                         #Checking if the result is found
            return result
    return None                                            #Returning None if the result is not found

print("All Paths:")
list_paths(entrance, [])

print("Path to Penguin Zone:")
path_to_penguin_zone = find_animal_path(entrance, "Penguin Zone", [])
print(path_to_penguin_zone)
```
