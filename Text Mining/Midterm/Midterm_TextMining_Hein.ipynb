## Part B — Python String Operations (30 points)

Use Python to clean the following text:
```
   Welcome to TEXT Mining 2026!!!
Visit: https://example.com
Contact: info@example.com
```

**Tasks**
- B1. Remove leading and trailing whitespace.(5 pts)  
- B2. Convert all characters to lowercase.(5 pts) 
- B3. Remove punctuation marks.(10 pts)
- B4. Remove extra spaces and line breaks so the text becomes a clean single line.(10 pts) 

Display output after each step.

````python
# Write your code for Part B here 

raw_text = """
   Welcome to TEXT Mining 2026!!!
Visit: https://example.com
Contact: info@example.com
"""

# B1. Remove leading and trailing whitespace
step_1 = raw_text.strip()
print("Step 1 (Strip):\n", step_1)

# B2. Convert to lowercase
step_2 = step_1.lower()
print("\nStep 2 (Lowercase):\n", step_2)

# B3. Remove punctuation marks
import string
# We use a translation table to remove all punctuation characters
step_3 = step_2.translate(str.maketrans('', '', string.punctuation))
print("\nStep 3 (No Punctuation):\n", step_3)

# B4. Remove extra spaces and line breaks (Single Line)
# split() without arguments handles all whitespace (tabs, newlines, spaces)
step_4 = " ".join(step_3.split())
print("\nStep 4 (Clean Single Line):\n", step_4)
````

## Part C — Regular Expressions & Text Cleaning (15 points)

Given the text:
```
Order ID: 45821, Date: 2026-03-08, Email: customer@test.com
```

**Tasks**
- C1. Write a regex pattern to extract all numbers.(5 pts)  
- C2. Remove the email address using regex substitution. (5 pts) 
- C3. Extract the date in YYYY-MM-DD format using regex. (5 pts) 

Show extracted results.

````python
# Write your code for Part C here

import re
#Given text
text_c = "Order ID: 45821, Date: 2026-03-08, Email: customer@test.com"

# C1. Extract all numbers
# \d+ matches one or more digits [cite: 356]
numbers = re.findall(r'\d+', text_c)
print("C1. Numbers extracted:", numbers)

# C2. Remove the email address
# Substitution replaces the pattern with an empty string [cite: 364]
no_email = re.sub(r'[\w\.-]+@[\w\.-]+', '', text_c)
print("C2. Text without email:", no_email.strip())

# C3. Extract the date in YYYY-MM-DD format
# Matches 4 digits, hyphen, 2 digits, hyphen, 2 digits
date_pattern = re.search(r'\d{4}-\d{2}-\d{2}', text_c)
if date_pattern:
    print("C3. Extracted Date:", date_pattern.group())
````

## Part D — Web Scraping Fundamentals (20 points)

Use Python with `requests` and `BeautifulSoup`.

**Tasks**
- D1. Write code to fetch a webpage from a given URL.(5 pts)  
- D2. Extract and display: page title, all h1 and h2 headings, and all hyperlinks (href). (10 pts) 
- D3. Add basic error handling to manage connection failures.(5 pts)

````python
# Write your code for Part D here

import requests
from bs4 import BeautifulSoup

url = "https://www.iana.org/domains/reserved"

try:
    # D1. Fetch the webpage
    response = requests.get(url, timeout=5)
    # Raise an error if the request was unsuccessful (e.g., 404 or 500)
    response.raise_for_status() 
    
    # D2. Parse and Extract
    soup = BeautifulSoup(response.text, 'html.parser')
    
    print("Page Title:", soup.title.text if soup.title else "No Title")
    
    print("\n--- H1 and H2 Headings ---")
    for heading in soup.find_all(['h1', 'h2']):
        print(f"{heading.name}: {heading.text.strip()}")
        
    print("\n--- Hyperlinks ---")
    for link in soup.find_all('a'):
        href = link.get('href')
        if href:
            print(href)

# D3. Basic Error Handling
except requests.exceptions.RequestException as e:
    print(f"Connection Error: {e}")
````

## Part E — Mini Applied Task: Text Cleaning Pipeline (25 points)

You are given raw HTML text:
```html
<html>
<body>
<h1>Text Mining Practice</h1>
<p>   This is a SAMPLE text!!! It contains numbers 12345 and symbols %$#.   </p>
</body>
</html>
```

**Tasks**
- E1. Extract visible text from the HTML. (5 pts)  
- E2. Clean the text by converting to lowercase, removing punctuation, and removing extra whitespace. (10 pts)
- E3. Present the final cleaned text. (5 pts) 
- E4. Briefly explain each preprocessing step and why it is important. (5 pts)

````python
# Write your code for Part E here

import re
html_content = """
<html>
<body>
<h1>Text Mining Practice</h1>
<p>   This is a SAMPLE text!!! It contains numbers 12345 and symbols %$#.   </p>
</body>
</html>
"""

# E1. Extract visible text
soup_e = BeautifulSoup(html_content, 'html.parser')
raw_visible_text = soup_e.get_text()

# E2 and E3. Cleaning Pipeline and Final Output
def clean_text_pipeline(text):
    # Convert to lowercase
    text = text.lower()
    # Remove punctuation using regex [cite: 393]
    text = re.sub(r'[^\w\s]', '', text)
    # Remove extra whitespace and join into single line
    text = " ".join(text.split())
    return text

final_cleaned_text = clean_text_pipeline(raw_visible_text)
print("Final Cleaned Text:", final_cleaned_text)
````

# E4. 
Explanation of Preprocessing Steps:

Lowercasing: Ensures that the same word (e.g., "Text" vs "text") is treated identically by the computer, reducing dimensionality.

Removing Punctuation: Punctuation marks usually do not contribute to the primary meaning for many text mining tasks and can be considered "noise".

Removing Extra Whitespace: Normalizes the text format so it is consistent and easier to tokenize or store in a structured database.
