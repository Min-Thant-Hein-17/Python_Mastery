```python
from PIL import Image

img = Image.open('C:\\Users\\ASUS\\Downloads\\Image.png')
img
```

```python
type(img)
```

```python
import numpy as np
img_array = np.array(img)
img_array
```

```python
img_array.shape
```

```python
# To use subplots, we need to import the subplots function from the matplotlib library
# Subplot! (not subplots in the code lines!)

#cmap = color map, it is a way to visualize the different channels of the image

import matplotlib.pyplot as plt

red = img_array[:, :, 0]
green = img_array[:, :, 1]
blue = img_array[:, :, 2]

plt.subplot(1, 3, 1)
plt.imshow(red, cmap='Reds')
plt.title('red channel')

plt.subplot(1, 3, 2)
plt.imshow(green, cmap='Greens')
plt.title('green channel')

plt.subplot(1, 3, 3)
plt.imshow(blue,cmap='Blues')
plt.title('blue channel')

plt.imshow(img_array)
```
