# Module 1: Introduction to Text Data

This notebook introduces text as a data type and explores
how text differs from structured numerical data.

```python
# Sample Text Data
text = """
Text mining is an interdisciplinary field that focuses on extracting
meaningful patterns and knowledge from unstructured text data.
"""
print(text)
```

```python
# Basic Text Properties
words = text.split()
sentences = text.split(".")

print("Number of words:", len(words))
print("Number of sentences:", len(sentences))
print("Unique words:", len(set(words)))
```
