Cconvert the following simple functions to lambda functions.

```python
def Get_First_Letter(name):
    """Take a name as input return the lowercase
    first letter of the name"""
    if name[0] == "J":
        first_letter = name[0].lower()
        return first_letter
    else:
        last_letter = name[-1]
        return last_letter


Get_First_Letter("Steve")
```

```python
#Define your lambda function here
f = lambda name: name[0].lower() if name[0] == "J" else name[-1]
f("Steve")
```

```python
def Check_Even(x):
    """Return True if x is even and odd otherwise """

    return x%2 == 0

Check_Even(5)
```

```python
#Define your lambda function here
g = lambda x: x%2 == 0
g(5)
```

```python
def Get_Num(L):
    """Return the integer version of the number
    specified by the list if strings in L"""

    return int("".join(L))

Get_Num(["1","5","2"])
```

```python
#Define your lambda function here
l = lambda L: int("".join(L))
l(["1","5","2"])
```

# Filtering in pandas

In this practice, we will explore data on the progression of salaries of students graduating from various schools. The file "salaries-by-region.csv" gives this information broken down by region and the the file "salaries-by-college-type.csv" gives this information broken by school type.

```python
import pandas as pd

def Get_Dict_Region(file_location):
    """Take as input the file location of salaries-by-region.csv
    and returns a dictionary with a key for each school whose value is the region
    of that school."""

    df_region = pd.read_csv(file_location)
    region_dict = {}
    for index, row in df_region.iterrows():
        region_dict[row['School Name']] = row['Region']
    return region_dict
```

Read in the file "salaries-by-college-type.csv" into a a dataframe called df_college_type and use dtype to check the how each column is stored. Use "N/A" as the na_values when you read in the dataframe.

```python
#Write your code here
df_college_type = pd.read_csv('/content/salaries-by-college-type.csv', na_values='N/A')
display(df_college_type.dtypes)
```

We need notice that all of the salary categories are objects when they should be ints. Use a for loop over the columns attribute of the dataframe df_college_type and an apply function to do the conversion of each of these columns to a float.

```python
#Write your code here
salary_columns = [
    'Starting Median Salary',
    'Mid-Career Median Salary',
    'Mid-Career 10th Percentile Salary',
    'Mid-Career 25th Percentile Salary',
    'Mid-Career 75th Percentile Salary',
    'Mid-Career 90th Percentile Salary'
]

for col in salary_columns:
    if col in df_college_type.columns:
        df_college_type[col] = df_college_type[col].astype(str).str.replace('$', '', regex=False).str.replace(',', '', regex=False).apply(lambda x: float(x) if x.strip() != '' else None)

display(df_college_type.dtypes)
display(df_college_type.head())
```

**Then** use map and a lambda function to add a column for region.

```python
#Write you code here
df_college_type['Region'] = df_college_type['School Name'].map(lambda x: D_regions.get(x))
display(df_college_type.head())
```

How many school-dept pairs have a starting salary greater than 65K?

```python
#Write your code here
salaries_above_65k = df_college_type[df_college_type['Starting Median Salary'] > 65000]
num_schools_above_65k = len(salaries_above_65k)
print(f"Number of school-dept pairs with a starting salary greater than 65K: {num_schools_above_65k}")
```

What is the average median starting salary among these schools?


```python
#Write your code here
average_starting_salary = salaries_above_65k['Starting Median Salary'].mean()
print(f"The average median starting salary among these schools is: ${average_starting_salary:,.2f}")
```

What is the median mid career 75th percentile salary of party schools in a Southern or Western region?

```python
#Filter for 'Party School' in 'School Type'
party_schools = df_college_type[df_college_type['School Type'] == 'Party School']

#Filter for Southern or Western regions
southern_western_party_schools = party_schools[party_schools['Region'].isin(['Southern', 'Western'])]

#Calculate the median mid-career 75th percentile salary
median_salary = southern_western_party_schools['Mid-Career 75th Percentile Salary'].median()

print(f"The median mid-career 75th percentile salary of party schools in a Southern or Western region is: ${median_salary:,.2f}")
```

What fraction of California schools have a starting median starting salary above 50K?

```python
# Filter for California schools
california_schools = df_college_type[df_college_type['Region'] == 'California']

# Filter California schools with starting median salary above 50K
salaries_above_50k_california = california_schools[california_schools['Starting Median Salary'] > 50000]

# Calculate the fraction
if not california_schools.empty:
    fraction = len(salaries_above_50k_california) / len(california_schools)
    print(f"Fraction of California schools with a starting median salary above 50K: {fraction:.2f}")
else:
    print("No California schools found in the dataset.")
```

How many schools have university in their name?

```python
num_university_schools = df_college_type[df_college_type['School Name'].str.contains('University', case=False, na=False)].shape[0]
print(f"Number of schools with 'University' in their name: {num_university_schools}")
```
