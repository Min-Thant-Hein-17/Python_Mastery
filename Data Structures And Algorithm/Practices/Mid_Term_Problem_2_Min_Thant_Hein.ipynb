```python
class QueueUsingStacks:
    def __init__(self):
        self.enqueue_stack = [] #Creating a stack to add elements
        self.dequeue_stack = [] #Creating a stack to remove elements

    def enqueue(self, x: int) :
        self.enqueue_stack.append(x) #Inserting item x to the queue

    def dequeue(self) :
        if not self.dequeue_stack:   #Checking whether the value is already in dequeue list or not
            while self.enqueue_stack:
                self.dequeue_stack.append(self.enqueue_stack.pop()) #Moving items from the enqueue stack to the dequeue stack
        if self.dequeue_stack:
            return self.dequeue_stack.pop() #Returning the item
        else:
          return None


    def peek(self) :
        if not self.dequeue_stack:  # To check the dequeue_stack is empty or not
            while self.enqueue_stack: #If dequeue_stack is empty, it will iterates until enqueue_stack has elements
                self.dequeue_stack.append(self.enqueue_stack.pop())  #to move elements from enqueue_stack to queue_stack
        if not self.dequeue_stack: #Checking whether dequeue_stack is empty or not
            return None   #dequeue_stack is empty
        else:
            return self.dequeue_stack[-1] #If it is not empty, it return the element at the last position

    def is_empty(self) -> bool: #To return either True or False
        return not self.enqueue_stack and not self.dequeue_stack #"not" operator is used to check whether self.enqueue_stack and self.dequeue_stack are empty
```

```python
#Example_Usage
q = QueueUsingStacks()
q.enqueue(1)
q.enqueue(2)
print(q.dequeue())  # Output: 1
print(q.peek())  # Output: 2
print(q.is_empty())  # Output: False
```
