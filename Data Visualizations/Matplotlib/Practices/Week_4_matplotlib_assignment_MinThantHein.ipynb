# Mini Assignment (20 minutes)

Make a “publication-ready” figure:

1. Plot **sin(x)** and **cos(x)** on the same axes
2. Add:
   - title
   - x/y labels
   - legend
   - grid
3. Zoom to x=0..6
4. Use custom ticks
5. Save as:
   - `final.png` at **300 dpi**
   - `final.pdf`
   - `final.svg`

> When finished, download the files from the `outputs/` folder.


```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# STARTER CELL for the mini project
fig, ax = plt.subplots(figsize=(7, 4))

# 1) Data
x = np.linspace(0, 10, 300)

# 2) Lines
ax.plot(x, np.sin(x), label="sin(x)")
ax.plot(x, np.cos(x), label="cos(x)")

# 3) Labels / title / legend / grid
ax.set_title("Final figure")
ax.set_xlabel("x")
ax.set_ylabel("value")
ax.grid(True, alpha=0.3)
ax.legend()

# 4) Zoom + ticks
ax.set_xlim(0, 6)
ax.set_xticks([0, 1, 2, 3, 4, 5, 6])

# 5) Save
import os
out_dir = "outputs"
os.makedirs(out_dir, exist_ok=True)

fig.savefig(os.path.join(out_dir, "final.png"), dpi=300, bbox_inches="tight")
fig.savefig(os.path.join(out_dir, "final.pdf"), bbox_inches="tight")
fig.savefig(os.path.join(out_dir, "final.svg"), bbox_inches="tight")

plt.show()
```
