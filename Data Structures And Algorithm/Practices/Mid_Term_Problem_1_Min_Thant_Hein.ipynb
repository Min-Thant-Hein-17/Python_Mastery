```python
import random

class RandomizedSet:
    def __init__(self):
      self.items = []  # To store elements
      self.dic = {}    #To store elements as keys and their corresponding indices in the list, items

    def insert(self, val: int) :
      self.dic[val] = len(self.items)
      self.items.append(val) # Adding value
      return val not in self.dic

    def remove(self, val: int):
      if val not in self.dic:
        return False                    # if value is not in the dictionary, retrun False
      index_to_remove = self.dic[val]   # Setting the index of the value to remove
      last_element = self.items[-1]     # Setting the last element in the list
      self.items[index_to_remove] = last_element  # changing the element that we want to remove with the last element
      self.dic[last_element] = index_to_remove    # removing the last element from list
      self.items.pop()
      del self.dic[val]  # removing the value from the dictionary
      return True

    def get_random(self):
      return random.choice(self.items)  #to select random element
```

```python
ds = RandomizedSet()
ds.insert(5)
ds.insert(10)
ds.remove(5)
print(ds.get_random())  # Should return 10
```
