## Step 3 — Create the Streamlit app

```python
%%writefile app.py
import streamlit as st
import pandas as pd
import matplotlib.pyplot as plt

st.title("Sales Dashboard (Critique Example)")
st.write("This dashboard intentionally contains design problems for critique.")

df = pd.read_csv("business_sales_dataset.csv")

df["date"] = pd.to_datetime(df["date"])
df["month"] = df["date"].dt.to_period("M").astype(str)

st.subheader("Revenue by Product")
fig, ax = plt.subplots()
prod_rev = df.groupby("product")["revenue"].sum()
ax.pie(prod_rev, labels=prod_rev.index, autopct="%1.1f%%")
st.pyplot(fig)

st.subheader("Revenue by Region")
fig, ax = plt.subplots()
region_rev = df.groupby("region")["revenue"].sum()
ax.pie(region_rev, labels=region_rev.index, autopct="%1.1f%%")
st.pyplot(fig)

st.subheader("Daily Revenue Trend")
fig, ax = plt.subplots()
daily_rev = df.groupby("date")["revenue"].sum()
ax.plot(daily_rev.index, daily_rev.values)
st.pyplot(fig)

st.subheader("Revenue by City")
fig, ax = plt.subplots()
city_rev = df.groupby("city")["revenue"].sum()
ax.bar(city_rev.index, city_rev.values)
plt.xticks(rotation=75)
st.pyplot(fig)

st.subheader("Discount vs Revenue")
fig, ax = plt.subplots()
ax.scatter(df["discount"], df["revenue"])
st.pyplot(fig)

st.subheader("Unit Price vs Units Sold")
fig, ax = plt.subplots()
ax.scatter(df["unit_price"], df["units_sold"])
st.pyplot(fig)
```
